```python
#42. feladat:

D = nx.DiGraph()
dependencies = {}

with open('day_21.txt', 'r') as f:
    inpt = f.read().split('\n')
    for item in inpt:
        item = item.split(' ')
        node = item[0][:-1]
        D.add_node(node)
        if len(item) == 2:
            D.nodes[node]['value'] = int(item[1])
            dependencies[node] = int(item[1])
        else:
            v1,v2 = item[1], item[3]
            D.add_edges_from([(v1, node), (v2, node)])
            dependencies[node] = item[1:]
            

wrong = nx.shortest_path(D, 'humn', 'root')[::-1]

new_deps = dependencies.copy()

D.remove_nodes_from(wrong[:-1])

top_order = list(nx.topological_sort(D))


for item in top_order:
    if type(dependencies[item]) != int:
        val_1, op, val_2 = dependencies[dependencies[item][0]], dependencies[item][1], dependencies[dependencies[item][2]]
        
        if op == '+':
            new_val = val_1 + val_2
        elif op == '-':
            new_val = val_1 - val_2
        elif op == '*':
            new_val = val_1 * val_2
        else:
            new_val = val_1 / val_2
        
        dependencies[item] = new_val


for i in range(len(wrong)-1):
    current_item = wrong[::-1][i]
    next_item = wrong[::-1][i+1]
    
    old_dep = dependencies[next_item]

    
    if old_dep.index(current_item) == 0:
        if old_dep[1] == '+':
            dependencies[current_item] = [next_item, '-', old_dep[2]]
        elif old_dep[1] == '-':
            dependencies[current_item] = [next_item, '+', old_dep[2]]
        elif old_dep[1] == '*':
            dependencies[current_item] = [next_item, '/', old_dep[2]]
        elif old_dep[1] == '/':
            dependencies[current_item] = [next_item, '*', old_dep[2]]
            
    elif old_dep.index(current_item) == 2:
        if old_dep[1] == '+':
            dependencies[current_item] = [next_item, '-', old_dep[0]]
        elif old_dep[1] == '-':
            dependencies[current_item] = [old_dep[0], '-', next_item]
        elif old_dep[1] == '*':
            dependencies[current_item] = [next_item, '/', old_dep[0]]
        elif old_dep[1] == '/':
            dependencies[current_item] = [old_dep[0], '/', next_item]
            
if dependencies['root'].index(wrong[1]) == 0:
    dependencies['root'] = dependencies[dependencies['root'][2]]

else:
    dependencies['root'] = dependencies[dependencies['root'][0]]
            
dependencies[wrong[1]] = dependencies['root'] 


for item in wrong[2:]:

    val_1, op, val_2 = dependencies[dependencies[item][0]], dependencies[item][1], dependencies[dependencies[item][2]]
        
    if op == '+':
        new_val = val_1 + val_2
    elif op == '-':
        new_val = val_1 - val_2
    elif op == '*':
        new_val = val_1 * val_2
    else:
        new_val = val_1 / val_2
        
    dependencies[item] = new_val


print(int(dependencies['humn']))
```
