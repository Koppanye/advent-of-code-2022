```python
import numpy as np

with open('day_24.txt', 'r') as f:
    inpt = f.read().split('\n')
    size_x = len(inpt)
    size_y = len(inpt[0])
    
    grid = np.array([item[i] for item in inpt for i in range(size_y)]).reshape((size_x, size_y))
    grid = grid[1:-1,1:-1]
    
    up = np.zeros_like(grid, dtype = 'int')
    down = np.zeros_like(grid, dtype = 'int')
    left = np.zeros_like(grid, dtype = 'int')
    right = np.zeros_like(grid, dtype = 'int')
    
    up[grid == '^'] = 1
    down[grid == 'v'] = 1
    left[grid == '<'] = 1
    right[grid == '>'] = 1
    

    
    
    stop = False
    cnt = 0
    previous_positions = [[-1,0]]

    
    while stop == False:
        up = np.vstack((up[1:,], up[0,].reshape((1,-1))))
        down = np.vstack((down[-1].reshape((1,-1)), down[:-1,]))
        left = np.hstack((left[:,1:], left[:,0].reshape((-1,1))))
        right = np.hstack((right[:,-1].reshape((-1,1)), right[:,:-1]))
        
        occupied = up + down + left + right
        
        
        potential_positions = [[x,y] for [x,y] in previous_positions] 
        potential_positions += [[x-1,y] for [x,y] in previous_positions if x-1 >= 0 and [x,y] != [-1,0]]
        potential_positions += [[x,y-1] for [x,y] in previous_positions if y-1 >= 0 and [x,y] != [-1,0]]
        potential_positions += [[x,y+1] for [x,y] in previous_positions if y+1 <= size_y-3 and [x,y] != [-1,0]]
        potential_positions += [[x+1,y] for [x,y] in previous_positions if x+1 <= size_x-3]
        
        potential = np.ones_like(occupied, dtype = 'int')
        potential[[x for [x,y] in potential_positions],[y for [x,y] in potential_positions]] = 0

        temp1 = list(np.where(potential + occupied == 0)[0])
        temp2 = list(np.where(potential + occupied == 0)[1])
        previous_positions= [[temp1[i], temp2[i]] for i in range(len(temp1))]
        #print(previous_positions)
        
        if [size_x-3, size_y-3] in previous_positions:
            stop = True
        cnt += 1
print(cnt + 1)


up = np.vstack((up[1:,], up[0,].reshape((1,-1))))
down = np.vstack((down[-1].reshape((1,-1)), down[:-1,]))
left = np.hstack((left[:,1:], left[:,0].reshape((-1,1))))
right = np.hstack((right[:,-1].reshape((-1,1)), right[:,:-1]))



up,down = np.rot90(down, k=2), np.rot90(up, k = 2)
left, right = np.rot90(right, k = 2), np.rot90(left, k = 2)

#print(up)
#print('')
#print(down)
#print('')
#print(left)
#print('')
#print(right)
#print('')
#print('')

stop = False
cnt = 0
previous_positions = [[-1,0]]

    
while stop == False:
    up = np.vstack((up[1:,], up[0,].reshape((1,-1))))
    down = np.vstack((down[-1].reshape((1,-1)), down[:-1,]))
    left = np.hstack((left[:,1:], left[:,0].reshape((-1,1))))
    right = np.hstack((right[:,-1].reshape((-1,1)), right[:,:-1]))
        
    occupied = up + down + left + right
        
        
    potential_positions = [[x,y] for [x,y] in previous_positions] 
    potential_positions += [[x-1,y] for [x,y] in previous_positions if x-1 >= 0 and [x,y] != [-1,0]]
    potential_positions += [[x,y-1] for [x,y] in previous_positions if y-1 >= 0 and [x,y] != [-1,0]]
    potential_positions += [[x,y+1] for [x,y] in previous_positions if y+1 <= size_y-3 and [x,y] != [-1,0]]
    potential_positions += [[x+1,y] for [x,y] in previous_positions if x+1 <= size_x-3]
        
    potential = np.ones_like(occupied, dtype = 'int')
    potential[[x for [x,y] in potential_positions if [x,y] != [-1,0]],[y for [x,y] in potential_positions if [x,y] != [-1,0]]] = 0

    temp1 = list(np.where(potential + occupied == 0)[0])
    temp2 = list(np.where(potential + occupied == 0)[1])
    previous_positions= [[temp1[i], temp2[i]] for i in range(len(temp1))] + [[-1,0]]

        
    if [size_x-3, size_y-3] in previous_positions:
        stop = True
    cnt += 1

print(cnt + 1)


up = np.vstack((up[1:,], up[0,].reshape((1,-1))))
down = np.vstack((down[-1].reshape((1,-1)), down[:-1,]))
left = np.hstack((left[:,1:], left[:,0].reshape((-1,1))))
right = np.hstack((right[:,-1].reshape((-1,1)), right[:,:-1]))



up,down = np.rot90(down, k=2), np.rot90(up, k = 2)
left, right = np.rot90(right, k = 2), np.rot90(left, k = 2)


stop = False
cnt = 0
previous_positions = [[-1,0]]

    
while stop == False:
    up = np.vstack((up[1:,], up[0,].reshape((1,-1))))
    down = np.vstack((down[-1].reshape((1,-1)), down[:-1,]))
    left = np.hstack((left[:,1:], left[:,0].reshape((-1,1))))
    right = np.hstack((right[:,-1].reshape((-1,1)), right[:,:-1]))
        
    occupied = up + down + left + right
        
        
    potential_positions = [[x,y] for [x,y] in previous_positions] 
    potential_positions += [[x-1,y] for [x,y] in previous_positions if x-1 >= 0 and [x,y] != [-1,0]]
    potential_positions += [[x,y-1] for [x,y] in previous_positions if y-1 >= 0 and [x,y] != [-1,0]]
    potential_positions += [[x,y+1] for [x,y] in previous_positions if y+1 <= size_y-3 and [x,y] != [-1,0]]
    potential_positions += [[x+1,y] for [x,y] in previous_positions if x+1 <= size_x-3]
        
    potential = np.ones_like(occupied, dtype = 'int')
    potential[[x for [x,y] in potential_positions if [x,y] != [-1,0]],[y for [x,y] in potential_positions if [x,y] != [-1,0]]] = 0

    temp1 = list(np.where(potential + occupied == 0)[0])
    temp2 = list(np.where(potential + occupied == 0)[1])
    previous_positions= [[temp1[i], temp2[i]] for i in range(len(temp1))] + [[-1,0]]

        
    if [size_x-3, size_y-3] in previous_positions:
        stop = True
    cnt += 1

print(cnt + 1)
```
