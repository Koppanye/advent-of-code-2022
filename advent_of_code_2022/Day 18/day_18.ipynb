```python
#35. feladat:

from itertools import product

with open('day_18.txt', 'r') as f:
    cubes = f.read().split('\n')
    cubes = [[int(i) for i in cube.split(',')] for cube in cubes]

    area = 0
    for cube in cubes:
        x, y, z = cube[0], cube[1], cube[2]
        neighbours = [[x,y,z-1], [x,y,z+1], [x, y-1, z], [x, y+1, z], [x-1, y, z], [x+1, y, z]]
        plus = [v for v in neighbours if v not in cubes]
        area += len(plus)

    print(area)
```

```python
#2. mo nx-szel

import networkx as nx
G = nx.Graph()
G.add_nodes_from([i for i in range(len(cubes))])

for v in G.nodes():
    G.nodes[v]['coords'] = cubes[v]
    

for u in G.nodes():
    for v in G.nodes():
        l1 = G.nodes[u]['coords']
        l2 = G.nodes[v]['coords']
        if len([i for i in range(3) if l1[i] != l2[i]]) == 1 and abs(sum(l1)-sum(l2)) == 1:
            G.add_edge(u,v)
    
```
